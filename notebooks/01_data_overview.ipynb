# 01 Data Overview

## Purpose
This notebook performs the initial inspection of the raw employee dataset.

The goal is to understand:
- the overall dataset shape
- the column structure
- data types
- missing values
- which columns look analytical versus administrative

## 1. Preview the dataset

First, I load the raw CSV and inspect the first few rows to understand the basic structure and the kinds of fields included.

```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
df = pd.read_csv('Messy_Employee_dataset.csv')
```

```python
df.head()
```

## 2. Dataset shape

I check the number of rows and columns to understand the size of the dataset.

```python
df.shape
```

## 3. Column names

This helps identify which fields may be useful for analysis and which may be metadata, identifiers, or contact-related columns.

```python
df.columns.tolist()
```

## 4. Data types

Reviewing data types is important because some columns may need conversion before analysis.
For example, date columns are often stored as text in raw CSV files.

```python
df.dtypes
```

## 5. Missing values

I count missing values in each column to identify where cleaning decisions will be needed.

```python
missing_counts = df.isna().sum().sort_values(ascending=False)
missing_counts[missing_counts > 0]
```

## 6. Missing value percentages

Percentages make it easier to judge whether missingness is minor, moderate, or substantial.

```python
missing_pct = (df.isna().mean() * 100).sort_values(ascending=False).round(2)
missing_pct[missing_pct > 0]
```

## 7. Basic schema summary

I separate columns into numeric, categorical, boolean, and other types for an early schema-level view.

```python
numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
bool_cols = df.select_dtypes(include=["bool"]).columns.tolist()

print("Numeric columns:")
print(numeric_cols)
print()

print("Categorical / object columns:")
print(categorical_cols)
print()

print("Boolean columns:")
print(bool_cols)
```

## 8. Quick uniqueness check

This helps identify columns that may behave like identifiers, names, emails, or phone numbers rather than reusable analytical variables.

```python
uniqueness_summary = pd.DataFrame({"n_unique" : df.nunique(dropna=False),
                                   "dype" : df.dtypes.astype(str),
                                   "missing_count" : df.isna().sum()
                                   }).sort_values("n_unique", ascending=False)
uniqueness_summary
```

## 9. Analytical vs administrative assessment

This is an early judgment step, not a final decision.
Some columns may be temporarily retained during cleaning even if they are not useful for later analysis.

### Likely analytical columns
- Age
- Salary
- Department_Region
- Join_Date
- Remote_Work

### Likely administrative / identifier / low-value columns
- First_Name
- Last_Name
- Email
- Phone

### Notes
- `Department_Region` looks like a compound field and may need to be split later.
- `Join_Date` may need datetime conversion.
- Contact-style columns may be retained temporarily for inspection, then excluded from the final analysis-ready dataset.

## 10. Initial observations

### Three-line summary
1. The dataset contains a mixed schema with numeric, categorical, boolean, and likely administrative fields.
2. Missing values appear to be especially important in `Age` and `Salary`, so those columns will require explicit cleaning decisions.
3. Some fields such as `Join_Date` and `Department_Region` suggest structural preprocessing will be needed before analysis.

## Why this step matters
Initial inspection comes before cleaning because preprocessing decisions should be based on the actual structure of the raw data, not assumptions.
By checking shape, dtypes, missingness, and column roles first, I can make later cleaning steps more reproducible and easier to justify.
